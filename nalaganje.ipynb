Sedaj iz csv datoteke s podatki preberemo url-naslove receptov in prenesemo še html dadoteke teh strani.

```python
import pandas as pd
import requests
import os
from tqdm import tqdm
```

Najprej odstranimo ' '.

```python
naslovi_receptov = pd.read_csv('podatki_z_indeksi.csv')
naslovi_receptov['povezava'] = naslovi_receptov['povezava'].map(lambda s: s[1:-1])
```

Sedaj shranimo podatke v datoteke "html_strani/indeks_recepta.html".

```python
for vrstica in tqdm(naslovi_receptov.iloc, total=len(naslovi_receptov)):
    id = vrstica['Unnamed: 0']
    pot = f"html_strani/{id}.html"
    if os.path.exists(pot):
        continue
    try:
        resp = requests.get(f"https://www.kulinarika.net{vrstica['povezava']}")
        with open(pot, 'w', encoding='utf-8') as file_out:
            file_out.write(resp.text)
    except Exception as e:
        print(f"Napaka pri prenosu: {vrstica['povezava']} ({id})")
```
